```python
data1 = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, 27, 22, 32, 29],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Salary': [70000, 80000, 65000, 95000, 74000],
    'Department': ['HR', 'Engineering', 'Sales', 'Marketing', 'Finance']
}

df = pd.DataFrame(data1)

data2 = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],  # Notice 'Eve' instead of 'Eva'
    'Gender': ['F', 'M', 'M', 'M', 'F'],
    'Experience (Years)': [3, 5, 2, 8, 4]
}
df1= pd.DataFrame(data2)
```

```python
inner_merged=pd.merge(df,df1, on='Name', how='inner')
```

```python
outer_merged=pd.merge(df,df1, on='Name', how='outer')
```

```python
# Sample data
data = {
    'Region': ['East', 'West', 'East', 'West', 'East', 'West'],
    'Product': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Sales': [250, 300, 400, 350, 500, 450]
}

df = pd.DataFrame(data)

df.pivot_table(index='Region', columns='Product', values='Sales')
```
